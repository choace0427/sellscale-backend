```python
import csv
import time

SOURCE_OF_TRUTH_COLUMN = "Email"
EMAILED = "Emailed?"
OPENED = "Opened?"
CLICKED = "Clicked?"
REPLIED = "Replied?"

FINISHED = "Finished"

st = time.process_time()

with open('outreach_ingestion_example.csv', newline='') as f:
    reader = csv.reader(f)
    
    column_names = next(reader)
    if SOURCE_OF_TRUTH_COLUMN not in column_names:
        #TODO: fix
        exit("No email column found")
    email_index = column_names.index(SOURCE_OF_TRUTH_COLUMN)
    
    sequence_state_index = column_names.index("Sequence State")
    name_index = column_names.index("Name")
    emailed_index = column_names.index(EMAILED)
    opened_index = column_names.index(OPENED)
    clicked_index = column_names.index(CLICKED)
    replied_index = column_names.index(REPLIED)

    for row in reader:
        email = row[email_index]
        sequence_state = row[sequence_state_index]
        if sequence_state != FINISHED:
            continue
        name = row[name_index]
        message = ""
        if row[replied_index] == "Yes":
            message = "has replied."
        elif row[clicked_index] == "Yes":
            message = "has clicked."
        elif row[opened_index] == "Yes":
            message = "has opened."
        elif row[emailed_index] == "Yes":
            message = "has been emailed."
        print(f"{name} ({email}) {message}")

print(time.process_time() - st)
```
