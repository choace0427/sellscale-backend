```python
from dotenv import load_dotenv
from pathlib import Path
import os
import sys
import json
import requests
import pandas as pd

env_path = Path('../..') / '.env'
load_dotenv(dotenv_path=env_path)

print(os.getcwd())
sys.path.append('Get from above^')

ISCRAPER_API_KEY = os.environ.get('ISCRAPER_API_KEY')
PROFILE_DETAILS_URL = "https://api.proapis.com/iscraper/v4/profile-details"

        
```

```python
def call_iscraper(linkedin_url: str):
    linkedin_id = linkedin_url.split("/in/")[1]
    payload = json.dumps(
        {
            "profile_id": linkedin_id,
            "profile_type": "personal",
            "network_info": True,
        }
    )
    headers = {"X-API-KEY": ISCRAPER_API_KEY, "Content-Type": "application/json"}

    response = requests.request(
        "POST", PROFILE_DETAILS_URL, headers=headers, data=payload
    )
    data = json.loads(response.text)

    return data
```

```python
payload = call_iscraper('linkedin.com/in/deepti-mishra-md-34144436')
state = payload['location']['state'] or payload['location']['default']
print(state)

```

```python
curative_csv = pd.read_csv('curative.csv')
for i in curative_csv.iterrows():
    if str(i[1]['State']) != 'nan' or str(i[1]['linkedin_url']) == 'nan':
        continue
    print(i[1]['linkedin_url'])
    payload = call_iscraper(i[1]['linkedin_url'])
    state = payload['location']['state'] or payload['location']['default']
    print(state)

    # write the state back to the csv in the State column
    curative_csv.loc[i[0], 'State'] = state
    curative_csv.to_csv('curative.csv', index=False)
```
